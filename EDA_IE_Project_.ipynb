```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
df
```

```python
Numerical_Features = df.select_dtypes(include=['int64', 'float64']).columns
Numerical_Features = Numerical_Features.drop('Class')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Numerical_Features' contains your selected numerical features
for feature in Numerical_Features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True)  # kde=True adds a kernel density estimate
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import pandas as pd

# Pearson correlation
pearson_corr = df[Numerical_Features].corr(method='pearson')

# Spearman correlation
spearman_corr = df[Numerical_Features].corr(method='spearman')

# Display the correlation matrices (you might want to use heatmaps for visualization)
print("Pearson Correlation:\n", pearson_corr)
print("\nSpearman Correlation:\n", spearman_corr)


# You can also create heatmaps for better visualization:
plt.figure(figsize=(12, 10))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Matrix')
plt.show()

plt.figure(figsize=(12, 10))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Spearman Correlation Matrix')
plt.show()
```

```python
# Compute correlations
pearson_corr = df.corr(method='pearson')
spearman_corr = df.corr(method='spearman')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 8))

sns.heatmap(pearson_corr, ax=axs[0], cmap='coolwarm', center=0, annot=False)
axs[0].set_title('Pearson Correlation')

sns.heatmap(spearman_corr, ax=axs[1], cmap='coolwarm', center=0, annot=False)
axs[1].set_title('Spearman Correlation')

plt.tight_layout()
plt.show()
```

```python
diff_corr = pearson_corr - spearman_corr

plt.figure(figsize=(12, 10))
sns.heatmap(diff_corr, cmap='RdBu', center=0, annot=False)
plt.title('Difference between Pearson and Spearman Correlations')
plt.show()
```
